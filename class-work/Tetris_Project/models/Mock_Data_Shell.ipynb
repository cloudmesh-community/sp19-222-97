```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import csv
import numpy as np
import matplotlib.pyplot as plt

#sort_nam = lambda e : e[0].sort()
#sort_scr = lambda e : e[1].sort()
#sort_lvl = lambda e : e[2].sort()

with open('./data/MOCK_DATA.csv') as csv_file:
    entries = []
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        print(row)
        if line_count == 0:
            #print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            #print(f'\tName: {row[0]} Score: {row[1]} Level: {row[2]} Playtime: {row[3]}')
            new_entry = row
            entries.append(new_entry)
            line_count += 1

    #sort_nam(entries)

    #entries_np_arr = np.asarray(entries)
    #np.matrix.sort(entries_np_arr)
    #print(entries)

    print(f'Processed {line_count} lines.')
```

```python
scores = []
playtimes = []
for e in entries:
    scores.append(e[1])
    playtimes.append(e[3])


#scores.sort()
#playtimes.sort()

#Plots the first 50 instances out of 1000
plt.plot(playtimes[0:49], scores[0:49], '.')
plt.yscale('linear')
plt.xscale('linear')
plt.show()
```

```python
from scipy import stats as st

X = np.array(playtimes[0:49]).astype(np.int)
Y = np.array(scores[0:49]).astype(np.int)

slope, intercept, r_val, p_val, std_err = st.linregress(X, Y)

plt.plot(X, Y, '.')
#plt.plot(X, slope*X + intercept, '-')
plt.yscale('linear')
plt.xscale('linear')
plt.show()

print(slope)
print(r_val)
```
